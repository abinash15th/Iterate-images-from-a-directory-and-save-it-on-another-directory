```python
import glob
from PIL import Image
i=1
images = glob.glob("C:/Users/user/abinash_all/*.jpg")
for image in images:
    with open(image,"rb") as file:
        img = Image.open(file)
        imgResult = img.resize((800,800), resample = Image.BILINEAR)
        imgResult.save("C:/Users/user/abi/"+str(i)+".jpg", 'JPEG')
        i=i+1
        print("All good")
```
